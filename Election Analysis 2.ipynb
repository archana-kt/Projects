## 1) Who secured the highest percentage of votes among all candidates?

```python
data.columns
```

```python
data['DMK Votes %']
```

```python
data['DMK Votes %'] = data['DMK Votes %'].str.rstrip('%').astype(float)
```

```python
data['DMK Votes %']
```

```python
highest_dmk_votes_candidate = data.loc[data['DMK Votes %'].idxmax()]
```

```python
highest_dmk_votes_candidate[['Candidate','DMK Votes %']]
```

```python
# So velusamy is having highest voting percentage
```

## 2) Did candidates from any specific constituency tend to perform better in terms of vote percentage?

```python
data.columns
```

```python
data.head(3)
```

```python
g = data.groupby(['PC Name','Candidate'])['DMK Votes %']
```

```python
constituency_avg_vote_percentage = g.mean().reset_index()
```

```python
constituency_avg_vote_percentage
```

```python
constituency_highest_avg_vote_percentage = constituency_avg_vote_percentage.loc[constituency_avg_vote_percentage['DMK Votes %'].idxmax()]
```

```python
print(constituency_highest_avg_vote_percentage[['Candidate','PC Name', 'DMK Votes %']])
```

## Both are same

## 3) Are there any noticeable trends or patterns in the vote percentages of candidates across different constituencies?

```python
data.columns
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.figure(figsize = (12, 6))
sns.boxplot(data = data, x = 'PC Name', y = 'DMK Votes %')
plt.title('Distribution of DMK Vote Percentages across Constituencies')
plt.xlabel('Constituency')
plt.ylabel('DMK Vote Percentage')
plt.xticks(rotation=90)
plt.show()
```
